# 04 — Cross-Pipeline Validation: Mouse Brain 1.3M

Compares all available pipelines. Runs on CPU.

**Note:** For 1.3M cells, standardized DE recomputation will be slow (~30-60 min per pipeline).

```python
import subprocess, sys, time
t0 = time.time()
result = subprocess.run(
    [sys.executable, "validate_cross_pipeline_harmonized.py",
     "--dataset", "mouse_brain_1m"],
    capture_output=True, text=True,
)
print(result.stdout)
if result.stderr:
    print("STDERR:", result.stderr[-2000:])
print(f"\nTotal validation time: {time.time()-t0:.1f}s ({(time.time()-t0)/60:.1f} min)")
```

## Quick runtime comparison

```python
import json, glob

spec_files = sorted(glob.glob("write/mouse_1m_*_spec.json"))
for sf in spec_files:
    with open(sf) as f:
        spec = json.load(f)
    pipeline = spec.get("pipeline", "unknown")
    timings = spec.get("timings_seconds", {})
    total = sum(timings.values()) if timings else 0
    print(f"\n{pipeline}:")
    for step, t in timings.items():
        print(f"  {step:20s}: {t:8.1f}s")
    print(f"  {'TOTAL':20s}: {total:8.1f}s ({total/60:.1f} min)")
```
